# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
#1. Create a subplot layout to accomodate different types of plotrs
plt.figure(figsize=(30,20))
```

```python
#1. Scatter Plot
plt.subplot(3,2,1)
plt.scatter(df['age'],df['chol'])
plt.title('Age vs Cholesterol')

# 2. Histogram
plt.subplot(3,2,2)
plt.hist(df['trestbps'],bins = 20)
plt.title('Max Heart Rate')

## Box plt
plt.subplot(3,2,3)
plt.hist(df['thalach'])
plt.title('Max Heart Rate')

# 4. Line Plot
plt.subplot(3,2,4)
df_sorted = df.sort_values(by='age') # sort by age for the line plot
plt.plot(df_sorted['age'],df_sorted['chol'],linestyle='--',color = 'b')
plt.title('Age Vs Cholesterol line')
#5. Bar plot
plt.subplot(3,2,5)
df['gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")

#6 Pie plot
plt.subplot(3,2,6)
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Heart disease Target Prodcution')
```

```python
plt.tight_layout()
plt.show()
```
