# Simple Random Sampling

```python
import pandas as pd
```

```python
df_2015 = pd.read_csv("marathon_results_2015.csv")
df_2016 = pd.read_csv("marathon_results_2016.csv")
df_2017 = pd.read_csv("marathon_results_2017.csv")
```

```python
df = pd.concat([df_2015,df_2016,df_2017])
```

```python
#get total number of rows in original data
original_row_count = len(df)
```

```python
#Perform simple random sampling (selecting  1/10th of the records )
sample_df = df.sample(frac=0.1,random_state=42)
```

```python
#Get the total number of rows in sampled dataframe
sample_row_count = len(sample_df)
```

```python
df.head()
```

```python
sample_df.head()
```

```python
print(f"number of rows in original dataframe:{original_row_count}")
print(f"number of rows in sample dataframe:{sample_row_count}")
```

# Stratified Random Sampling

```python
#Find the total number of males and females athelete
total_males = len(df[df['M/F'] == 'M'])
total_females = len(df[df['M/F'] == 'F'])
```

```python
total_males
```

```python
total_females
```

```python
#Calculate the proportionate number of males and female athelete for a sample
sample_size = int(0.1*original_row_count)
print(sample_size)
sample_males = int(0.1* total_males)
print(sample_males)
sample_females = int(0.1*total_females)
print(sample_females)
```

```python
## Generate proportationate samples of male and female
male_Samples = df[df['M/F'] == 'M'].sample(n=sample_males,random_state=42)
male_Samples
```

```python
female_samples = df[df['M/F'] == 'F'].sample(n= sample_females,random_state=42)
female_samples
```

```python
##Combining the male and female samples into a single dataframe
sample_df = pd.concat([male_Samples,female_samples])
sample_row_count = len(sample_df)
```

# Cluster Sampling

```python
#Find unique contries the athletes are from
unique_countries = df['Country'].unique()
num_countries= len(unique_countries)
num_countries
```

```python
#Randomly select half of countries
import numpy as np
np.random.seed(42)
```

```python
selected_countries = np.random.choice(unique_countries, size=num_countries//2, replace=False)
selected_countries
```

```python
#Create a new dataframe containing all records of selcted countries
sample_df = df[df['Country'].isin(selected_countries)]
```

```python
#Get total number of rows in sample dataframe
sample_row_count= len(sample_df)
sample_row_count
```
