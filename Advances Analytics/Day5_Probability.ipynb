```python
#For the marathon dataset find the conditional probability to finish within 3 hours given that pace is less than 8 minutes per mile/km
```

```python
import pandas as pd

df_2015 = pd.read_csv("marathon_results_2015.csv")
df_2016 = pd.read_csv("marathon_results_2016.csv")
df_2017 = pd.read_csv("marathon_results_2017.csv")
```

```python
df = pd.concat([df_2015,df_2016,df_2017])
```

```python
df[['hour','mins','seconds']] = df['Official Time'].str.split(':', expand = True)

df['hour'] = pd.to_numeric(df['hour'])
#df['Hour'] = df['Hour'].astype(int)
df['mins'] = pd.to_numeric(df['mins'])
df['seconds'] = pd.to_numeric(df['seconds'])

df['finish_time_in_seconds'] = df['hour']*3600 + df['mins']*60 + df['seconds']
```

```python
df.drop(['hour','mins','seconds','Total_time'], axis=1,inplace=True)
df
```

```python
df[['hour','mins','seconds']] = df['Pace'].str.split(':', expand = True)

df['hour'] = pd.to_numeric(df['hour'])
#df['Hour'] = df['Hour'].astype(int)
df['mins'] = pd.to_numeric(df['mins'])
df['seconds'] = pd.to_numeric(df['seconds'])

df['pace_in_seconds'] = df['hour']*3600 + df['mins']*60 + df['seconds']
```

```python
#Define conditions
condition_A = df['finish_time_in_seconds'] <= 3*60*60   #Finish within 3 hours
condition_B = df['pace_in_seconds'] <= 8 *60 # Pace less than 8 minutes per KM
```

```python
#Calculate probalities
P_B = len(df[condition_B])/ len(df)
P_A_and_B = len(df[condition_A & condition_B])/len(df) 
```

```python
#Conditional probability P(A|B)
P_A_given_B = P_A_and_B/P_B
```

```python
print(f"P(A | B )= {P_A_given_B: .4f}")
```
