# Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
#Set seaborn style
sns.set(style='darkgrid')
```

```python
#Create a sublot layout to accomodate diiferent types of plot
plt.figure(figsize=(16,14))
```

```python
#1. Sactter plot
plt.figure(figsize=(16,14))
plt.subplot(3,3,1)
df['gender_mod'] = df['gender'].map({0:'Female',1:'Male'})
sns.scatterplot(data = df,x='age',y='chol',hue = 'gender_mod',alpha = 0.7)
plt.title('Age vs Cholesterol')
plt.legend(title='gender',loc='upper left',bbox_to_anchor=(1,1))
```

```python
# 2 Histogram with KDE
plt.figure(figsize=(16,14))
plt.subplot(3,3,2)
sns.histplot(df['trestbps'],kde = True,color = 'blue')
plt.title('Resting BP Distribution')
```

```python
# Box plot
plt.figure(figsize=(16,14))
plt.subplot(3,3,3)
sns.boxplot(data=df,y='thalach',x='target',palette='Set2')
plt.title('Max Heart Rate by Target')
```

```python
#4 Line plot
plt.figure(figsize=(16,14))
plt.subplot(3,3,4)
sns.lineplot(data=df,x='age',y='chol',color = 'orange')
plt.title('Age vs cholesterol')
```

```python
# 5. Bar plot (improved count plot for 'cp' features)
plt.figure(figsize=(16,14))
plt.subplot(3,3,5)
sns.countplot(data=df,x='cp',palette='Set1',hue = 'target')
plt.title('Chest Pain Type Distribution by Target')
```

```python
#6 Heatmap (coorelation matrix)
plt.figure(figsize=(16,14))
plt.subplot(3,3,6)
relevant_columns= ['age','chol','thalach','trestbps','oldpeak','target']
correlation_matrix = df[relevant_columns].corr()
sns.heatmap(correlation_matrix,annot = True,cmap = 'coolwarm',fmt = '.2f',square = True,linewidths = 0.5)
plt.title('Heatmap of selected Variables')
```

```python
plt.tight_layout()
plt.sho
```

```python
##pair plot
sns.pairplot(df[relevant_columns], hue= 'target', palette='husl', corner=True, diag_kind='kde')
plt.suptitle('Pair plot of Selected Features', y=1.02)
plt.show()
```
