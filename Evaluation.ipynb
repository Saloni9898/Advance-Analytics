```python
df_data1 = df_data.groupby(['Innings Player','Country'])
```

```python
df_y = df_data1.sum()
```

```python
sample_India = df_India.sample(n=30,random_state =42)
```

```python
sample_Australia = df_Australia.sample(n=30,random_state =42)
```

```python
sample_India
```

```python
sample_Australia
```

```python
Bowling_average1 = sample_India['Innings Runs Conceded']/sample_India['Innings Wickets Taken']
```

```python
Bowling_average2 = sample_Australia['Innings Runs Conceded']/sample_Australia['Innings Wickets Taken']
```

```python
Bowling_average1
```

```python
Bowling_average2
```

```python
# Perform an independent two-sample t-test
t_stat, p_value = stats.ttest_ind(Bowling_average1, Bowling_average2)
```

```python
# Print the result
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
alpha = 0.05 
```

```python
if p_value < alpha:
    print("Reject the null hypothesis: The means are significantly different.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the means.")
```
